##Expasión de celulas madre mesenquimales en condiciones hipóxicas en medio no xenógenico suplementado con plasma pobre en plaquetas.

grupo de investigación  Medicina Regenerativa

Universidad Autónoma de Bucaramanga (UNAB)

linea de acción: Medicina, Biología y salud (Ingeniería de tejidos y terapia celular de células madres)

Elaboración de cultivos de vintro con condiciones hipóxicas suplementado con plasma pobre en plaquetas (PPP) que permite la la expanción de las MSC, esto se aplicaria para en la medicina regenerativa
ellos quieres utilizar la implementación de las células madre mesenquimales que hacen parte del desarrollo de los innovadores distintos métodos de la regeneración de tejidos, permitiendo fabricar innovadoras alternativas para la medicina regenerativa. su objetivo de evaluar el efecto de las condiciones hipóxicas y medio no xenogénico en la expansión de MSC. y a su vez buscan establecer el tiempo de duplicación poblacional de las MSC cultivadas en condiciones hipóxicas y medio de cultivo suplementado con ppp.
